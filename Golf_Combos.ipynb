```python
import pandas as pd
import itertools
import numpy as np
```

```python
golf_df = pd.read_csv('sentry.csv')
golf_df.columns
```

```python
golf_df = golf_df.loc[:, ['Player','Salary', 'Own', 'My Proj']]
golf_df['Own'] = golf_df['Own'].str.replace('%','').astype(float)
golf_df.head()
```

```python
combos = list(itertools.combinations(golf_df['Player'], 6))
print(combos[:5])
print(len(combos))
```

```python
combos_df = pd.DataFrame(combos, columns=['G1','G2','G3','G4','G5','G6'])
combos_df.head()
```

```python
projection_map = dict(zip(golf_df['Player'], golf_df['My Proj']))
ownership_map = dict(zip(golf_df['Player'], golf_df['Own']))
salary_map = dict(zip(golf_df['Player'], golf_df['Salary']))
```

```python
proj_cols = []
own_cols = []
salary_cols = []
for i in range(1,7):
  combos_df[f'G{i}_Proj'] = combos_df[f'G{i}'].map(projection_map)
  proj_cols.append(f'G{i}_Proj')
for i in range(1, 7):
  combos_df[f'G{i}_Own'] = combos_df[f'G{i}'].map(ownership_map)
  own_cols.append(f'G{i}_Own')
for i in range(1, 7):
  combos_df[f'G{i}_Salary'] = combos_df[f'G{i}'].map(salary_map)
  salary_cols.append(f'G{i}_Salary')
```

```python
combos_df.head()
```

```python
combos_df['Total Proj'] = combos_df[proj_cols].sum(axis=1)
combos_df['Own Sum'] = combos_df[own_cols].sum(axis=1)
combos_df['Salary Cap'] = combos_df[salary_cols].sum(axis=1)
combos_df.head()
```

```python
print(len(combos_df))
combos_df = combos_df[combos_df['Salary Cap'] <= 50000]
print(len(combos_df))
combos_df.head()
```

```python
clean_combos = combos_df.loc[:, ['G1','G2','G3','G4','G5','G6','G1_Own','G2_Own','G3_Own','G4_Own','G5_Own','G6_Own','Total Proj','Own Sum','Salary Cap']]
clean_combos.head()
```

```python
clean_combos = clean_combos.sort_values(by=['Total Proj'], ascending = False)
clean_combos.head()
```

```python
clean_combos[:40000].to_csv('sentry_lineups.csv')
```
