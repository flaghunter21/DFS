```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import itertools
```

```python
hitters_df = pd.read_csv('hitters.csv')
hitters_df['Own%'] = hitters_df['Own%'].str.replace('%','').astype(float)
hitters_df.head()
```

```python
cubs_df = hitters_df[hitters_df['Team'] == 'CHC']
cubs = list(itertools.combinations(cubs_df['Player'], 5))
cubs_combos = pd.DataFrame(cubs)
pirates_df = hitters_df[hitters_df['Team'] == 'PIT']
pirates = list(itertools.combinations(pirates_df['Player'], 5))
pirates_combos = pd.DataFrame(pirates)
tigers_df = hitters_df[hitters_df['Team'] == 'DET']
tigers = list(itertools.combinations(tigers_df['Player'], 5))
tigers_combos = pd.DataFrame(tigers)
redsox_df = hitters_df[hitters_df['Team'] == 'BOS']
redsox = list(itertools.combinations(redsox_df['Player'], 5))
redsox_combos = pd.DataFrame(redsox)
yankees_df = hitters_df[hitters_df['Team'] == 'NYY']
yankees = list(itertools.combinations(yankees_df['Player'], 5))
yankees_combos = pd.DataFrame(yankees)
rays_df = hitters_df[hitters_df['Team'] == 'TB']
rays = list(itertools.combinations(rays_df['Player'], 5))
rays_combos = pd.DataFrame(rays)
giants_df = hitters_df[hitters_df['Team'] == 'SF']
giants = list(itertools.combinations(giants_df['Player'], 5))
giants_combos = pd.DataFrame(giants)
braves_df = hitters_df[hitters_df['Team'] == 'ATL']
braves = list(itertools.combinations(braves_df['Player'], 5))
braves_combos = pd.DataFrame(braves)
bluejays_df = hitters_df[hitters_df['Team'] == 'TOR']
bluejays = list(itertools.combinations(bluejays_df['Player'], 5))
bluejays_combos = pd.DataFrame(bluejays)
whitesox_df = hitters_df[hitters_df['Team'] == 'CWS']
whitesox = list(itertools.combinations(whitesox_df['Player'], 5))
whitesox_combos = pd.DataFrame(whitesox)
cardinals_df = hitters_df[hitters_df['Team'] == 'STL']
cardinals = list(itertools.combinations(cardinals_df['Player'], 5))
cardinals_combos = pd.DataFrame(cardinals)
brewers_df = hitters_df[hitters_df['Team'] == 'MIL']
brewers = list(itertools.combinations(brewers_df['Player'], 5))
brewers_combos = pd.DataFrame(brewers)
royals_df = hitters_df[hitters_df['Team'] == 'KC']
royals = list(itertools.combinations(royals_df['Player'], 5))
royals_combos = pd.DataFrame(royals)
angels_df = hitters_df[hitters_df['Team'] == 'LAA']
angels = list(itertools.combinations(angels_df['Player'], 5))
angels_combos = pd.DataFrame(angels)
dbacks_df = hitters_df[hitters_df['Team'] == 'ARI']
dbacks = list(itertools.combinations(dbacks_df['Player'], 5))
dbacks_combos = pd.DataFrame(dbacks)
padres_df = hitters_df[hitters_df['Team'] == 'SD']
padres = list(itertools.combinations(padres_df['Player'], 5))
padres_combos = pd.DataFrame(padres)
```

```python
combos = pd.concat([cubs_combos, pirates_combos, tigers_combos, redsox_combos, yankees_combos, rays_combos, giants_combos, braves_combos, bluejays_combos, whitesox_combos, cardinals_combos, brewers_combos, royals_combos, angels_combos, dbacks_combos, padres_combos], axis=0)
print(len(combos))
combos.columns = ['H1','H2','H3','H4','H5']
combos.head()
```

```python
projection_map = dict(zip(hitters_df['Player'], hitters_df['My Proj']))
score_map = dict(zip(hitters_df['Player'], hitters_df['Actual Score']))
ownership_map = dict(zip(hitters_df['Player'], hitters_df['Own%']))
salary_map = dict(zip(hitters_df['Player'], hitters_df['Salary']))
team_map = dict(zip(hitters_df['Player'], hitters_df['Team']))
```

```python
proj_cols = []
own_cols = []
score_cols = []
salary_cols = []
for i in range(1,6):
  combos[f'H{i}_Proj'] = combos[f'H{i}'].map(projection_map)
  proj_cols.append(f'H{i}_Proj')
for i in range(1, 6):
  combos[f'H{i}_Own'] = combos[f'H{i}'].map(ownership_map)
  own_cols.append(f'H{i}_Own')
for i in range(1, 6):
  combos[f'H{i}_Score'] = combos[f'H{i}'].map(score_map)
  score_cols.append(f'H{i}_Score')
for i in range(1, 6):
  combos[f'H{i}_Salary'] = combos[f'H{i}'].map(salary_map)
  salary_cols.append(f'H{i}_Salary')
for i in range(1, 6):
  combos[f'H{i}_Team'] = combos[f'H{i}'].map(team_map)
  salary_cols.append(f'H{i}_Team')
```

```python
combos.head()
```

```python
combos['Stack Proj'] = combos[proj_cols].sum(axis=1)
combos['Stack Own%'] = combos[own_cols].sum(axis=1)
combos['Stack Score'] = combos[score_cols].sum(axis=1)
combos['Stack Salary'] = combos[salary_cols].sum(axis=1)
combos['Stack Team'] = combos['H1_Team']
combos.head()
```

```python
clean_combos = combos.loc[:,['H1','H2','H3','H4','H5','Stack Team','Stack Proj','Stack Score','Stack Own%', 'Stack Salary']]
clean_combos.head()
```

```python
clean_combos.to_csv('clean_combos.csv',index=False)
```
